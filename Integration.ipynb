Now, one way to approximate the integral of f(x) with respect to x from x=a to x=b is by using Reinman's sum. This technque can be described intuitively as followed: 

*   Divide the interval [a,b] in n partitions ( for convinience lets make all the n paritions equal width). Lets say $x_i$ is the beginning of the $i^{th}$ sub interval, each sub-interval is Δx width long. 

*  Calculate the following sum: $\sum_{i=0}^{n}f(x_i)Δx$

If Δx -> 0, then this sum is equal to the integral $\int_{a}^{b}f(x)dx$

**Task 1**: 
Write code for calculating the integral of a function using the Reinman's sum. 

* Calcualte the value of  $\int_1^{10} x^2 dx$  using this function  (take number of intervals = 4)
* Set the number of intervals as n= 4,5,100,1000 and calculate the relative errors. 





```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def riemann_sum(func, n, a, b):
    # func is the function that will be used to calculate f(x)
    # n = number of partitions in the interval
    # a = start of the interval
    # b = end of the interva;
    sum = None 

    # hint: use a for loop to claculate a sum. For calculating this sum, find the initial point of the subinterval x_i, pass it to function(), multiply the returned value with x_i and add to the sum. 
    
    dx = (b - a) / n
    f = sp.lambdify(x, func)

    sum = 0.0
    for i in range(n):
      sum += f((dx * i) + a) * dx

    return sum

x = sp.Symbol('x')
y = x**2

#value of riemann sum 
print(riemann_sum(y, 4, 1, 10))

intervals = [4, 5, 100, 1000]

appx_of_riemann_sum = []
errors_of_appx = []

for n in intervals:
  appx_of_riemann_sum.append(riemann_sum(y, n, 1, 10))

#calculating relative errors
for i in range(len(appx_of_riemann_sum) - 1):
    error = (abs(appx_of_riemann_sum[i+1] - appx_of_riemann_sum[i])) / appx_of_riemann_sum[i+1]
    errors_of_appx.append(error * 100)

plt.figure(figsize=(10,6))
plt.plot([1, 2, 3], errors_of_appx[:], 'r*-')
plt.title('Relative Errors')
plt.xlabel('Number of iterations')
plt.ylabel('Approximate Riemann Sum')
plt.grid('both')


```

###Observation###

Error decreases as soon as we start to take smaller intervals i.e. increase the total number of intervals.

Another aproach of calcualting intergals is the trapizoidal rule. Simply put, this rule approcimates the original function using a straight line. This transformes the area under the curve to a trapizoid. More simply put, using trapizoidal rule, 

$\int_a^b f(x)dx = \frac{(b-a)}{2}(f(a)+f(b))$

```python
def trapezoid_sum(func, a, b):
    # func is the function that will be used to calculate f(x)
    # a = start of the interval
    # b = end of the interva;
    sum = None 

    f = sp.lambdify(x, func)
    sum = (b - a) * (f(a) + f(b)) / 2        

    return sum

print(trapezoid_sum(y, 1, 10))
```
