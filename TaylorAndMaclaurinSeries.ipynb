# Solution to Task 01

```
import math

#Setting up the initial running parameters. 
max_iter = 50
x = 2.5
```

```
def exp_mclauren(x=0, n=1):
    """
    returns the summation of the first n terms of the McLauren series for e^x considering:
    x: exponent of the natural number i.e. Euler's number(e).
    n: number of terms to consider from the expansion series. 
    """
    series_sum = 0 
    for i in range(0, n):
        series_sum += math.pow(x, i) / math.factorial(i)
    # print(series_sum)
    return series_sum
```

```
# Calculating approximation
result_array = []

# Calculating the sum of the series with different values of n to compare. 
for i in range(0, max_iter):
    result_array.append(exp_mclauren(x=x, n=i))

## Calculating relative approximate errors
errors = []
for i in range(1, len(result_array)):
    e = abs(result_array[i] - result_array[i-1]) *100 / result_array[i]
    errors.append(e)
```

```
# plotting the results to analyze. 
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(result_array[:], 'g*-')
plt.title('Calculation of of e^x Numerically \n from McLauren Series')
plt.xlabel('Number of terms considered')
plt.ylabel('value of e^x')
plt.grid('both')
plt.show()


plt.figure(figsize=(10,6))
plt.plot(errors[:], 'b*-')
plt.title('Relative approximate errors in approximating e^x')
plt.xlabel('Iterations')
plt.ylabel('Absolute approximate error values')
plt.grid('both')
plt.show()
```

Observatons:


1.   In the first figure, we can see that the value is gaining more accuracy per iteration but the change begins to narrow down after a certain number of iterations. In the 3rd iteration it was 3.5 in the 2nd iteration and 6.6 after the 3rd but after 13 iterations it's less than $10^{-5}$
2.   In the second figure, error starts at 74% but by 15 iterations error reduces well below $10^{-8}$
3.   From above two points we can say that the errors are reducing at an exponential rate. It's because of the function being exponential. Maclaurin series is an exponential function. It gains acccuracy at an exponential rate thus reducing approximate errors exponentially.



#Solution to Task 02

```
def exp_mclauren_sine(x=0, n=1):
    """
    returns the summation of the first n terms of the McLauren series for sin(x) considering:
    x: measure of the angle
    n: number of terms to consider from the expansion series. 
    """
    series_sum = 0 
    for i in range(0, n):
        j = (2*i) + 1
        if (i == 0 or i%2 == 0):
            series_sum += math.pow(x, j) / math.factorial(j)
        else:
            series_sum -= math.pow(x, j) / math.factorial(j)
    return series_sum
```

```
# Calculating approximation
result_array_sine = []

# Calculating the sum of the series with different values of n to compare. 
for i in range(0, max_iter):
    result_array_sine.append(exp_mclauren_sine(x=x, n=i))

## Calculating relative approximate errors
errors_sine = []
for i in range(1, len(result_array_sine)):
    e = abs(result_array_sine[i] - result_array_sine[i-1]) *100 / result_array_sine[i]
    errors_sine.append(e)
```

```
# plotting the results to analyze. 
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(result_array_sine[:], 'g*-')
plt.title('Calculation of of sin(x) Numerically \n from McLauren Series')
plt.xlabel('Number of terms considered')
plt.ylabel('value of sin(x)')
plt.grid('both')
plt.show()


plt.figure(figsize=(10,6))
# plt.plot(errors_sine, 'b*-')
plt.plot(errors_sine[:], 'b*-')
plt.title('Relative approximate errors in approximating sin(x)')
plt.xlabel('Iterations')
plt.ylabel('Absolute approximate error values')
plt.grid('both')
plt.show()
```

Observatons:


1.   In the first figure, we can see that the value is gaining more accuracy per iteration but the change begins to narrow down after a certain number of iterations. In the 3rd iteration it was 0.71 but after 6 iterations it's less than $10^{-5}$
2.   In the second figure, error for 2nd iteration is 118% and for 3rd iteration is 21% but by 8 iterations error reduces well below $10^{-6}$
3.   From above two points we can say that the errors are reducing at an exponential rate. It's because of the function being exponential. Maclaurin series is an exponential function. It gains acccuracy at an exponential rate thus reducing approximate errors exponentially.
